```python
def send_request(model, prompt):
    retry_count = 0
    prompt_parts = [prompt]

    while retry_count < MAX_RETRY:
        try:
            response = model.generate_content(prompt_parts)
            # print(response.text)
            return response.text
        except Exception as e:
            print(f"Error: {e}")
            time.sleep(RETRY_DELAY)
            retry_count += 1

    return None
```

```python
def get_prompt(question):
    return f""" Assume you are English expert. Now paraphrase the following Question while maintaining the meaning. Note each questions ask about a certain Disease and only give one paraphrased question.

    Few examples are provided with question and paraphrased question. Interpret them and follow the similar approach to rephrase the given question.

    Examples:

    Example1:
    Question: Causes such as Suddenly slowing down, changing direction, cutting, Pivoting with your foot firmly planted, Landing awkwardly from jump, Stopping suddenly, direct blow to the knee, having a collision could lead to which Disease?

    Paraphrased Question: What injury or condition can be caused by activities like abrupt stops, changes in direction, forceful pivoting, awkward landings, sudden impacts, or direct blows to the knee?

    Example2:
    Question: Preventions like Bath twice, avoid fatty spicy food, drink plenty of water, avoid too many products avoids which Disease ?

    Paraphrased Question: What condition can be prevented by practices like regular bathing, a balanced diet low in fatty and spicy foods, sufficient water intake, and limiting the use of certain products?

    Example3:
    Question: The factors such as If you travel to an area where mosquitoes continue to carry the yellow fever virus. have a risk of which Disease ?

    Paraphrased Question:  What disease is a risk if you travel to a region where mosquitoes are known to transmit the yellow fever virus?

    Example4:
    Question: Diagnosis such as Phonocardiography are used to identify which Disease ?

    Paraphrased Question: What condition is diagnosed using phonocardiography?

    Example5:
    Question: Symptoms such as burning micturition, bladder discomfort, foul smell of urine, continuous feel of urine indicate which Disease ?

    Paraphrased Question: What condition is characterized by symptoms like painful urination, bladder discomfort, foul-smelling urine, and a persistent sensation of needing to urinate?

    Example6:
    Question: Which Disease could lead to complications such as Hearing loss, Memory problems, Learning disabilities, Brain damage, Trouble walking, Seizures, Kidney failure, Shock, Death ?

    Paraphrased Question: What condition can result in serious complications such as hearing loss, memory problems, learning difficulties, brain damage, impaired mobility, seizures, kidney failure, shock, and even death?


    Your task:
    Paraphrase the given question, do it even if its present in the examples.
    
    Question: {question}

    Paraphrased Question: """
```
